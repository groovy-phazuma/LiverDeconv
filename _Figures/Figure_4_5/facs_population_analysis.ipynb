#### FACS population analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***
##### 1st trial: 2022/07/13 - 2022/07/13

```python
#%% 1st
# add ctrl information
facs = pd.read_csv('C:/github/LiverDeconv/_Figures/Figure_4_5/data/FACS_1st.csv',index_col=0)
tmp = facs.T
tmp.insert(0,'APAP#0hr',1)
facs = tmp.T
c = 'tab:blue'
row_n=2
col_n=6
cells = facs.columns.tolist()
fig = plt.figure(figsize=(18*row_n,2*col_n),dpi=300)
for i,cell in enumerate(cells):
    target_df = facs[[cell]]
    df = target_df.loc[['APAP#12hr','APAP#24hr','APAP#48hr']]
    df['variable'] = [t.split("#")[1] for t in df.index.tolist()]
    df.columns = ['value','variable']
    ax = fig.add_subplot(row_n,col_n,i+1) # identify the plotting location
    # violin
    sns.violinplot(x='variable', y='value', data=df, inner=None, cut=0, scale="count",linewidth=1.5, palette="winter")
    plt.setp(ax.collections, alpha=.6)
    # strip
    sns.stripplot(x='variable', y='value', data=df, jitter=True, linewidth=1.5, size=9, dodge=True, palette="winter")
    # ctrl base line
    plt.hlines(y=1,xmin=-1.5,xmax=3-0.5,color="red",ls="dashed",linewidth=2,zorder=-2)
    
    # mean value plotting
    df2 = target_df.loc[['APAP#0hr','APAP#12hr','APAP#24hr','APAP#48hr']]
    df2['variable'] = [t.split("#")[1] for t in df2.index.tolist()]
    df2.columns = ['value','variable']
    df_means = df2.groupby('variable').mean()['value'].tolist()
    plt.scatter(x=[-1],y=[1],marker="s",color='blue',s=150)
    plt.plot([i-1 for i in range(len(df_means))],df_means,linewidth=2,zorder=-1,alpha=0.9)
    
    ax.set_axisbelow(True)
    ax.grid(color="#ababab",linewidth=0.5)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().yaxis.set_ticks_position('left')
    plt.gca().xaxis.set_ticks_position('bottom')
    plt.title(cell,fontsize=15)
    plt.xlabel("Hours post acetaminophen",fontsize=15)
    plt.ylabel("deconvolution values",fontsize=15)
plt.show()
```
